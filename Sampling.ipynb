### Sampling techniques

Here are we going to use the following sampling techniques to aid with monte carlo simulation:
- Uniform sampling
- Sabol sequence

#### Uniform sampling

- The simplest sampling technique is uniform sampling. This is where we sample from a uniform distribution.
- If you look at the plot below, there are some areas that are not sampled. This is not ideal for monte carlo simulation as we want to sample from the entire space.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns

num_samples = 10_000

rs = np.random.RandomState(0)
uniform_samples = rs.rand(2, num_samples)

# PLot the uniform samples
plt.scatter(uniform_samples[0], uniform_samples[1], s=1)
```

#### Sobol sequence

- This is a quasi-random sequence/monte carlo method that is designed to fill the space more evenly.
- Used for integration, optimization and monte carlo simulation.

```python
sobol_sampler = st.qmc.Sobol(2, # number of dimensions 
                             scramble=False)    # scramble the position of rows 
sobol_samples = sobol_sampler.random(num_samples)

# Plot the sobol samples
plt.scatter(sobol_samples[:, 0], sobol_samples[:, 1], s=1)
```
